```python
# proportion of each value in Category column
discounts_df['Category'].value_counts(normalize=True)
```

```python
# filling missing values in Category column as per existing proportion
def fill_na_category_proportionally(dis_df, col_name):
    
    # calculating value counts and proportions of the non-NaN values
    value_counts = dis_df[col_name].value_counts(normalize=True)
    missing_count = dis_df[col_name].isnull().sum()

    if missing_count == 0:
        print(f"No missing values found in '{col_name}' column.")
        return dis_df

    # generating random values based on the calculated proportions
    fill_values = np.random.choice( value_counts.index, size=missing_count,
                                    p=value_counts.values)

    # indices of NaN values
    nan_indices = dis_df[dis_df[col_name].isnull()].index

    # filling the NaN values with the randomly generated values
    dis_df.loc[nan_indices, col_name] = fill_values

    print(f"Filled {missing_count} missing values in '{col_name}' column proportionally.")
    print(f"\nUpdated value proportion:\n{round(dis_df[col_name].value_counts(normalize=True),2)}")

    return dis_df

discounts_df = fill_na_category_proportionally(discounts_df, 'Category')
```

```python
# proportion of each value in Sub Category column
discounts_df['Sub Category'].value_counts(normalize=True)
```

```python
# filling missing values in Sub Category column as per existing proportion
def fill_na_sub_category_proportionally(dis_df, col_name):
    
    # calculating value counts and proportions of the non-NaN values
    value_counts = dis_df[col_name].value_counts(normalize=True)
    missing_count = dis_df[col_name].isnull().sum()

    if missing_count == 0:
        print(f"No missing values found in '{col_name}' column.")
        return dis_df

    # generating random values based on the calculated proportions
    fill_values = np.random.choice( value_counts.index, size=missing_count,
                                    p=value_counts.values)

    # indices of NaN values
    nan_indices = dis_df[dis_df[col_name].isnull()].index

    # filling the NaN values with the randomly generated values
    dis_df.loc[nan_indices, col_name] = fill_values

    print(f"Filled {missing_count} missing values in '{col_name}' column proportionally.")
    print(f"\nUpdated value proportion:\n{round(dis_df[col_name].value_counts(normalize=True),2)}")

    return dis_df

discounts_df = fill_na_sub_category_proportionally(discounts_df, 'Sub Category')
```

#### **3. Dealing with Outliers**

```python
transactions_df.head()
```

```python
# identifying Product ID wise outlier percentage for Unit Price column
 
outlier_summary = {}

for product_id, tempdf in transactions_df.groupby('Product ID')['Unit Price']:
    Q1 = tempdf.quantile(0.25)
    Q3 = tempdf.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_mask = (tempdf < lower_bound) | (tempdf > upper_bound)
    outlier_count = outliers_mask.sum()
    total_count = len(tempdf)

    outlier_percent = round((outlier_count / total_count) * 100, 2)

    outlier_summary[product_id] = {'Outlier Count': outlier_count, 'Outlier %': outlier_percent}

outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index').sort_values(by='Outlier %', ascending=False)

outlier_df.head(12)
```

<small>Detecting outliers in Unit Price column using IQR (Inter Quartile Range) method and replacing with upper and lower bound values by applying clipping on all Unit Price values within group formed by Product ID</small>

```python
transactions_df['Unit Price'] = transactions_df.groupby('Product ID')['Unit Price'].transform(lambda x: x.clip(
    lower=x.quantile(0.25) - 1.5 * (x.quantile(0.75) - x.quantile(0.25)),
    upper=x.quantile(0.75) + 1.5 * (x.quantile(0.75) - x.quantile(0.25))
))
```

```python
# updated outiler counts

outlier_summary = {}

for product_id, tempdf in transactions_df.groupby('Product ID')['Unit Price']:
    Q1 = tempdf.quantile(0.25)
    Q3 = tempdf.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_mask = (tempdf < lower_bound) | (tempdf > upper_bound)
    outlier_count = outliers_mask.sum()
    total_count = len(tempdf)

    outlier_percent = round((outlier_count / total_count) * 100, 2)

    outlier_summary[product_id] = {'Outlier Count': outlier_count, 'Outlier %': outlier_percent}

outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index').sort_values(by='Outlier %', ascending=False)

outlier_df.head(12)
```

<small>3.1) Product sub categories experience significant variations in Gross Margin based on seasons where the winter cloths peak in colder months and summer clothes dominate warmer months.<small>

```python
# grouping Gross Margin by Season & Sub Category
season_margin = zara_df.groupby(['Seasons', 'Sub Category'], observed=True)['Gross Margin'].sum().unstack()

season_margin = round((season_margin / 1_000_000), 2)
season_margin = season_margin.T

# segmenting winter and summer subcategories
winter_subcategories = ['Coats and Blazers', 'Sweaters and Knitwear', 'Sweaters and Sweatshirts', 
                        'Suits and Blazers', 'Pants and Jeans']

summer_subcategories = ['T-shirts and Tops', 'Skirts and Shorts', 'Dresses and Jumpsuits', 
                        'Sportswear', 'Accessories']

# filtering dataset to include only the selected subcategories
season_margin_filtered = season_margin.loc[winter_subcategories + summer_subcategories]

print("Gross Margin (in millions) by Season and Sub Category:")
season_margin_filtered
```

```python
# Define colors for different subcategories
subcategory_colors = {sub_cat: plt.cm.Set1(i / len(season_margin_filtered.index)) for i, sub_cat in enumerate(season_margin_filtered.index)}

# Define linestyle mapping for seasons (Winter = Dashed, Summer = Solid)
linestyle_map = {sub_cat: '--' if sub_cat in winter_subcategories else '-' for sub_cat in season_margin_filtered.index}

# Plot each subcategory separately while keeping all seasons on x-axis
plt.figure(figsize=(12, 6))

for sub_cat in season_margin_filtered.index:
    plt.plot(season_margin_filtered.columns, season_margin_filtered.loc[sub_cat], 
             marker='o', label=sub_cat, color=subcategory_colors[sub_cat], linestyle=linestyle_map[sub_cat])

# Create legend handles for Winter & Summer categories
winter_legend = mlines.Line2D([], [], color='black', linestyle='--', label="Winter Subcategories")
summer_legend = mlines.Line2D([], [], color='black', linestyle='-', label="Summer Subcategories")

# Formatting the plot
plt.title("Seasonal Impact on Gross Margin by Sub Category", fontsize=14, fontweight='bold')
plt.xlabel("Season", fontsize=12, fontweight='bold')
plt.ylabel("Gross Margin (Millions)", fontsize=12, fontweight='bold')
plt.legend(handles=[winter_legend, summer_legend] + [mlines.Line2D([], [], color=subcategory_colors[sub_cat], marker='o', linestyle=linestyle_map[sub_cat], label=sub_cat) for sub_cat in season_margin_filtered.index], bbox_to_anchor=(1.05, 1), loc='upper left', title="Sub Categories")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

<small>Winter clothings shows expected high performance in colder months.
- Coats and Blazers ($6.68M in Winter vs. $4.36M in Summer) 
- Sweaters and Knitwear ($2.07M in Winter vs. $1.46M in Summer) 
- Suits and Blazers ($4.91M in Winter vs. $3.17M in Summer) 
- Pants and Jeans ($6.61M in Winter vs. $4.16M in Summer)

This demonstrates a clear seasonality effect with peak demand in colder months, confirming the hypothesis, aligning with temperature based seasonal trend.

However, Summer the clothings donâ€™t show extreme seasonality and actually perform slightly better in non-summer seasons.

4) **Store Performance**

<small>4.1) Stores with more employees generate higher sales, leading to better customer service and improved purchase conversion rates.<small>

```python
# grouping stores by Number of Employees and computing Gross Margin to check correlation with store size
store_performance = zara_df.groupby('Number of Employees')['Gross Margin'].sum()

store_performance_million = store_performance / 1000000


emp_summary = zara_df['Number of Employees'].describe()
min_emp = emp_summary['min'] - 10
percentile_25_emp = emp_summary['25%']
percentile_50_emp = emp_summary['50%']
percentile_75_emp = emp_summary['75%']
max_emp = emp_summary['max'] + 10

# new column for Store Employee Count Group
zara_df['Store Emp Count Group'] = pd.cut(zara_df['Number of Employees'], 
                                          bins=[min_emp, percentile_25_emp, percentile_50_emp, max_emp], 
                                          labels=["Small", "Medium", "Large"])

store_performance_grouped = zara_df.groupby('Store Emp Count Group')['Gross Margin'].sum() / 1000000


fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Line Plot: Number of Employees vs. Gross Margin
sns.lineplot(x=store_performance.index, y=store_performance_million, color='darkblue', ax=axes[0])
axes[0].set_title("Store Performance: Employees vs. Gross Margin", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Number of Employees", fontsize=12, fontweight='bold')
axes[0].set_ylabel("Gross Margin (Millions)", fontsize=12, fontweight='bold')
axes[0].grid(True, linestyle='--', alpha=0.5)

# **Bar Chart: Employee Groups vs. Gross Margin**
sns.barplot(x=store_performance_grouped.index, y=store_performance_grouped.values, palette='viridis', ax=axes[1], edgecolor='black')

# Add value annotations above bars
for index, value in enumerate(store_performance_grouped.values):
    axes[1].text(index, value + 0.1, f"{value:.2f}", ha='center', fontsize=12, fontweight='bold', color='black')

axes[1].set_title("Gross Margin by Employee Count Group", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Store Employee Count Group", fontsize=12, fontweight='bold')
axes[1].set_ylabel("Gross Margin (Millions)", fontsize=12, fontweight='bold')
axes[1].grid(axis='y', linestyle='--', alpha=0.5)

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

<small>Positive Correlation between employee count and Gross Margin can be seen as the line chart shows a clear upward trend, indicating that as the number of employees increases, the gross margin also grows. This suggests that larger teams may contribute to stronger sales performance, possibly due to better customer service, faster operations, and improved store management.

The bar chart shows that stores with a small number of employees have the highest gross margin ($73.63M), followed by large stores ($64.99M) and then medium stores ($54.05M). This contradicts that more employees always lead to better performance which might indicate efficiency in smaller teams or differences in store types and customer engagement.

Hence, considering the correlation and increasing gross margin with increase in employee count we can conclude the hypothesis to be correct.

5) **Regional Demand**

<small>5.1) Stores in high-density urban areas generate higher sales volume and Gross Margin compared to stores in suburban or rural regions. <small>

```python
# store location segmentation based on city names global population density and economic influence 
# (took help from google search)

urban_cities = ['New York', 'Los Angeles', 'Chicago', 'Shanghai', 'Beijing', 'Shenzhen', 
                'Berlin', 'London', 'Paris', 'Madrid', 'Barcelona']

suburban_cities = ['Houston', 'Phoenix', 'Guangzhou', 'Chongqing', 'Hamburg', 'Birmingham', 
                   'Frankfurt am Main', 'Marseille', 'Lyon', 'Lisboa', 'Porto']

rural_cities = ['MÃ¼nchen', 'KÃ¶ln', 'Glasgow', 'Liverpool', 'Bristol', 'Toulouse', 'Nice', 
                'Sevilla', 'Zaragoza', 'Braga', 'GuimarÃ£es', 'Coimbra']

# segmenting stores into Urban, Suburban, and Rural
def categorize_city(city):
    if city in urban_cities:
        return "Urban"
    elif city in suburban_cities:
        return "Suburban"
    elif city in rural_cities:
        return "Rural"

zara_df['Store Location Type'] = zara_df['Store_City'].apply(categorize_city)

# grouping data by store type
store_performance = zara_df.groupby('Store Location Type')[['Gross Margin', 'Invoice Total']].sum()
store_performance[['Gross Margin', 'Invoice Total']] /= 1000000

# plotting grouped bar chart
fig, ax = plt.subplots(figsize=(6, 5))
store_performance.plot(kind='bar', ax=ax ,color=['#FF6347', "#2491EA"], width=0.9, edgecolor='black')

ax.set_title("Comparison of Gross Margin & Invoice Total by Store Type", fontsize=12, fontweight='bold')
ax.set_xlabel("Store Location Type", fontsize=10, fontweight='bold')
ax.set_ylabel("Amount (Millions)", fontsize=10, fontweight='bold')
ax.grid(axis='y', linestyle='--', alpha=0.5)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", fontsize=8, fontweight='bold')

plt.legend(title="Metrics")
plt.show()
```

<small>The grouped bar chart shows that urban locations generate the highest Gross Margin, confirming the hypothesis that high-density areas drive stronger revenue. This suggests that metropolitan areas contribute to greater profitability.

Suburban locations have lower Gross Margin than urban stores but still perform better than rural stores.

Rural locations consistently show lowest sales volume and Gross Margin confirming that lower population density struggle with fewer customers, leading to reduced foot traffic and purchasing power.
